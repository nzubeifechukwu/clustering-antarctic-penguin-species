```python
# Import Required Packages
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import silhouette_score
```

```python
# Loading and examining the dataset
penguins_df = pd.read_csv("penguins.csv")
penguins_df.head()
```

```python
X = penguins_df.drop("sex", axis="columns").values
```

```python
xs = X[:,0]
ys = X[:,1]
plt.scatter(xs, ys)
plt.show()
```

```python
xs = X[:,2]
ys = X[:,3]
plt.scatter(xs, ys)
plt.show()
```

```python
clusters = []
inertias = []
for c in range(1, 8):
    model = KMeans(n_clusters=c)
    model.fit(X)
    inertias.append(model.inertia_)
    clusters.append(c)
```

```python
plt.plot(clusters, inertias, marker="o")
plt.show()
```

```python
model = KMeans(n_clusters=4)
model.fit(X)
labels = model.labels_
```

```python
xs = X[:,0]
ys = X[:,1]
plt.scatter(xs, ys, c=labels)
plt.show()
```

```python
xs = X[:,2]
ys = X[:,3]
plt.scatter(xs, ys, c=labels)
plt.show()
```

```python
centroids = model.cluster_centers_
centroids_df = pd.DataFrame(centroids, columns=penguins_df.drop("sex", axis="columns").columns)
centroids_df
```

```python
labels = model.predict(X)
```

```python
xs = X[:,2]
ys = X[:,3]
plt.scatter(xs, ys, c=labels)
plt.show()
```

```python
scaler = StandardScaler()
scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
X_scaled
```

```python
xs = X_scaled[:,2]
ys = X_scaled[:,3]
plt.scatter(xs, ys, c=labels)
plt.show()
```

```python
centroids_df.plot(kind="bar")
```

```python
pipeline = make_pipeline(StandardScaler(), KMeans(n_clusters=4))
```

```python
pipeline.fit(X)
```

```python
labels = pipeline.named_steps["kmeans"].labels_
centroids = pipeline.named_steps["kmeans"].cluster_centers_
centroids
```

```python
pd.DataFrame(centroids, columns=penguins_df.drop("sex", axis="columns").columns)
```

```python
sex_dummies = pd.get_dummies(penguins_df["sex"])
```

```python
penguins = pd.concat([penguins_df.select_dtypes("number"), sex_dummies], axis="columns")
penguins.head()
```

```python
inertia_err = []
silhouette_scores = []
n_clusters = range(2, 13)
for i in n_clusters:
    model = make_pipeline(StandardScaler(), KMeans(n_clusters=i, random_state=42))
    model.fit(penguins)
    inertia_err.append(model.named_steps["kmeans"].inertia_)
    silhouette_scores.append(silhouette_score(penguins, model.named_steps["kmeans"].labels_))
```

```python
plt.plot(n_clusters, inertia_err, marker="o")
```

```python
plt.plot(n_clusters, silhouette_scores, marker="o")
```

```python
final_model = make_pipeline(StandardScaler(), KMeans(n_clusters=4, random_state=42))
final_model.fit(penguins)
```

```python
penguins_num = penguins_df.select_dtypes("number")
labels = final_model.named_steps["kmeans"].labels_
penguins_num["labels"] = labels
penguins_num.head()
```

```python
penguins_df.head()
```

```python
plt.scatter(penguins_df["culmen_length_mm"], penguins_df["culmen_depth_mm"], c=labels)
```

```python
plt.scatter(penguins_df["flipper_length_mm"], penguins_df["body_mass_g"], c=labels)
```

```python
penguins_gb = penguins_num.groupby("labels").mean()
penguins_gb
```

```python
penguins_gb[["culmen_length_mm", "culmen_depth_mm"]].plot(kind="bar")
```

```python
penguins_gb["flipper_length_mm"].plot(kind="bar")
```
